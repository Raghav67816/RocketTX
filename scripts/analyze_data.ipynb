### Analyze Data
**"log_data.py"** writes the loadcell values to your file which is generally names as your **Motor Name**

Use this jupyter notebook to get the average force value and for plotting post-test graphs.

```python
# Import dependencies
import tkinter.filedialog as tkf
# Open the file
file_dialog = tkf.askopenfile("r", defaultextension=".txt", title="Open Force Value File")
if file_dialog is None:
    pass
else:
    data = file_dialog.readlines()
```

```python
# Get the sum and number of observations
obs = len(data)
x = 0
num = 0
nums = []
print(f"Observations = {obs}")
for i in range(0, obs):
    num = data[i]
    x += int(num.split("\n")[0])
    nums.append(int(num.split("\n")[0]))
del data
print(f"Sum of observations: {x}")
print(f"Average Force: {x/obs} N")
```

### Plot graph for the values stored

```python
# Import dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_array = np.array(nums) # Y-axis (thrust)
time = 60
y = []
for i in range(0, len(nums)):
    y.append(i)
print(f"Length of nums: {len(nums)}")
print(f"Length of y: {len(y)}")
```

```python
plt.plot(y, nums)
plt.xlabel("Time (Secs)")
plt.ylabel("Thrust (N)")
plt.show()
```
